```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
```

```python
data=pd.read_csv("train.csv")
df_train=pd.DataFrame(data)
sur=pd.Series(df_train['Survived'])
df_train=df_train.drop("Survived",1)
df_train['Survived']=sur
df_train.shape
```

```python
data2=pd.read_csv("test.csv")
df_test=pd.DataFrame(data2)
data3=pd.read_csv("gender_submission.csv")
df3=pd.DataFrame(data3)
df_test=pd.merge(df_test,df3,on="PassengerId")
df_test.shape
```

```python
df=pd.concat([df_train,df_test])
df.shape
```

```python
df.head()
```

```python
#For identifying number of unique values
for i in df:
  print(i,":",df[i].nunique())
```

```python
df=df.drop("Cabin",1)
```

```python
df.isna().sum()
```

```python
df=df.drop("Cabin",1)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
df['Fare'].fillna(df['Fare'].mode()[0],inplace=True)
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
label_encode=preprocessing.LabelEncoder()
df['Sex']=label_encode.fit_transform(df['Sex'])
df['Embarked']=label_encode.fit_transform(df['Embarked'])
```

```python
df
```

```python
df.isna().sum()
```

```python
# Example: Countplot of survivors
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()

# Example: Relationship between class and survival
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Class')
plt.show()

# Example: Distribution of ages
sns.histplot(x=df['Age'].dropna(),y=df['Survived'].dropna())
plt.title('Survival by Age')
plt.show()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```
